```python
import json
import matplotlib.pyplot as plt
import glob

from os.path import join
from os import walk, path
```

```python
SRC_DIR = '../experiments/lr_search'
KEY = 'lr'
```

```python
def load_json(file_path):
    with open(file_path, 'r') as file:
        out = json.load(file)
    return out
```

```python
paths = glob.glob(join(SRC_DIR, f'{KEY}_*', 'log.json'))
acc_values = []
loss_vals = []

titles = []

for log_file in paths:
    data = load_json(log_file)
    acc_values.append(data['acc'])
    loss_vals.append(data['loss'])
    titles.append(path.basename(path.dirname(log_file)))
```

```python
len(acc_values)
```

```python
fig, ax = plt.subplots()

for i, acc in enumerate(acc_values):
    ax.plot(acc, label=titles[i])

ax.title.set_text('acc')
plt.legend(loc="upper left")
plt.show()


fig, ax = plt.subplots()

for i, loss in enumerate(loss_vals):
    ax.plot(loss, label=titles[i])
ax.title.set_text('loss')
plt.show()
```
